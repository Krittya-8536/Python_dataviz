# Basic Barchat (กราฟแท่งง่ายๆ)

### สร้างกราฟแท่งเปรียบเทียบรายได้ในปี 2541 ของจังหวัดกรุงเทพและปริมณฑล

```python
plt.bar([1,2,3,4],greaterbangkok .iloc[:,2])
plt.title('Barchart compareing average monthly incoms of provinces in Graeter Bangkok in 2541') #ตั้งชื่อกราฟ
plt.xlabel('Provinces')   #แกน x พร้อมตั้งชื่อในวงเล็บ
plt.ylabel('Income Baht/Month/Household')   #แกน y พร้อมตั้งชื่อในวงเล็บ
```

### เพิ่มชื่อให้แก่ละแกน คำสั่ง tick_label = ชื่อไฟล์.iloc[แถวที่เลือก,คอลัมที่เลือก]

```python
plt.bar([1,2,3,4],greaterbangkok .iloc[:,2],color='G',tick_label=greaterbangkok.iloc[:,-1])
plt.title('Barchart compareing average monthly incoms of provinces in Graeter Bangkok in 2541') #ตั้งชื่อกราฟ
plt.xlabel('Provinces')   #แกน x พร้อมตั้งชื่อในวงเล็บ
plt.ylabel('Income Baht/Month/Household')   #แกน y พร้อมตั้งชื่อในวงเล็บ
```

### compare different year กราฟเปรียบเทียบ 2 ปี
#### align='edge' กำหนดขอบ
#### plt.legend() กำหนดสีให้แต่ละปีในกราฟ

```python
plt.bar([1-(0.35/2),2-(0.35/2),3-(0.35/2),4-(0.35/2)],greaterbangkok.iloc[:,-2],width=0.35,tick_label=greaterbangkok.iloc[:,-1],color='G',label='2558',align='edge')
plt.bar([1+(0.35/2),2+(0.35/2),3+(0.35/2),4+(0.35/2)],greaterbangkok.iloc[:,2],width=0.35,tick_label=greaterbangkok.iloc[:,-1],color='b',label='2541',align='edge')
plt.title('Barchart compareing average monthly incoms of provinces in Graeter Bangkok in 2541,2558')
plt.xlabel('Provinces')   #แกน x พร้อมตั้งชื่อในวงเล็บ
plt.legend()
plt.ylabel('Income Baht/Month/Household')   #แกน y พร้อมตั้งชื่อในวงเล็บ
```

# #Stacked Barchat

```python
saledata=pd.DataFrame([{'North':20,'East':10,'West':90,'South':120},
                        {'North':220,'East':20,'West':50,'South':90}])
saledata
```

```python
plt.bar([1,2],saledata.iloc[:,0],tick_label= ['Company A','Company B'])
plt.bar([1,2],saledata.iloc[:,1],bottom=saledata.iloc[:,0],tick_label= ['Company A','Company B'])
plt.bar([1,2],saledata.iloc[:,2],bottom=saledata.iloc[:,0]+saledata.iloc[:,1],tick_label= ['Company A','Company B'])
plt.bar([1,2],saledata.iloc[:,3],bottom=saledata.iloc[:,0]+saledata.iloc[:,1]+saledata.iloc[:,2],tick_label= ['Company A','Company B'])
   
```
