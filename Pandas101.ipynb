```python
import pandas as pd
```

```python
w_resterant_PATH ='C:\Tangmo\Programming\Python_dataviz\w_restaurant.csv'
search_click_PATH='C:\Tangmo\Programming\Python_dataviz\search_click.csv'
w_resterant_df=pd.read_csv(w_resterant_PATH)
search_click_df=pd.read_csv(search_click_PATH)
```

```python
w_resterant_df.head()
```

## Select only some columm

```python
small_table=w_resterant_df[['name','city_name','zipcode','weighted_average_rating']]
small_table.head()
```

```python
search_click_df.head()
```

## add colun from another table การรวมตารางแบบตัดแปะ

```python
small_table['id']=w_resterant_df['id']
```

```python
small_table.head()
```

# .map()

# add business name to search_click

```python
search_click_df['business_name']= search_click_df['business_id'].map(small_table.set_index('id')['name'])
search_click_df.head()
```

# . merge() 

### combine two table รวมกันทั้งตาราง
#### how = 'left' บอกว่าตัวไหนคือตารางหลัก left_on ='business_id',right_on='id' บอกคีย์ของตารางที่ตรงกัน

```python
merged_table =search_click_df.merge(small_table, how = 'left',left_on ='business_id',right_on='id') 
```
